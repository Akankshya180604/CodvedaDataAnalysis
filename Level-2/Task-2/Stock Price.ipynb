# TASK-2:TIME SERIES ANALYSIS

## OBJECTIVE:
- Plot time-series data and identify patterns.
- Decompose the series into trend, seasonality, and residuals using statsmodels.
- Perform moving average smoothing and plot the results.

### 1. IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

### 2. READING THE CSV FILE

```python
stock=pd.read_csv('Stock Prices Data Set.csv')
print(stock.head())
```

```python
stock['date']=pd.to_datetime(stock['date'])
stock.set_index('date',inplace=True)
stock.head()
```

### 3. FILTER THE DATA FOR A SINGLE STOCK SYMBOL 

```python
symbol=stock['symbol'].unique()
print('Available symbols:',symbol)
```

```python
selected_symbol=symbol[0]
```

```python
symbol_data=stock[stock['symbol']==selected_symbol]['close']
```

### 4. PLOT TIME SERIES DATA

```python
plt.figure(figsize=(8,6))
plt.plot(symbol_data)
plt.title('Time Series Plot- Close price for symbol {selected_symbol}')
plt.xlabel('Date')
plt.ylabel('Standardize close price')
plt.show()
```

### 5. DECOMPOSE THE TIME SERIES

```python
res=seasonal_decompose(symbol_data,model='attitude',period=30)
res.plot()
plt.tight_layout()
plt.show()
```

### 6. MOVING AVERAGE SMOOTHING

```python
mov_avg=symbol_data.rolling(window=30).mean()
```

```python
plt.figure(figsize=(10,6))
plt.plot(symbol_data,label='Original')
plt.plot(mov_avg,label='30-day moving average',color='red')
plt.title(f'moving average smoothing  - Close price for symbol{selected_symbol}')
plt.xlabel('Date')
plt.ylabel('Standardize close price')
plt.legend()
plt.show()
```

#### INTERPRETATION:
- Plotted Time-Series of stock close price- observed fluctuations, growth phases and sharp declines.
#### Time series decomposition was performed into this following components. 
1. Trend: Reflected the long term upward or downward movement of data.
2. Seasonality: identified recurring patterns occuring at regular intervals(e.g. monthly or daily).
3. Residuals: represents the random variation left after removing trend and seasonality.

##### Applied moving average smoothing- minimized fluctuations, highlighted underlying trend and provided a clear view of overall price movement.
