### 1. IMPORTING LIBRARIES

### 2. READING THE CSV FILE

```python
churn=pd.read_csv('churn-bigml-80.csv')
print(churn.head())
```

### 5. VISUALIZATION

#### a)BAR PLOT

```python
plt.figure(figsize=(10,8))
sns.barplot(x=churn['Churn'],y=churn['Total day minutes'])
plt.title('Bar plot',fontweight='bold')
plt.xlabel('Churn status',color='blue')
plt.ylabel('Total Day Minutes',color='blue')
plt.legend(['Total Churn'],title='Legend')
plt.savefig('Bar plot for total day minutes .png',dpi=500,bbox_inches='tight')
plt.show()
```

#### INTERPRETATION:
- The box plot is used to show the distribution of Total day minutes by churn status.
- The maximum churned customers has total around 220 minutes a day.
- The maximum unchurned customers has total 175 minutes a day.
- From this graph it is cleared that customers having maximum Total Day Minutes are likely to churned. 

#### b)LINE CHART:

```python
plt.figure(figsize=(8,4))
gd=churn.groupby('Customer service calls')['Total day minutes'].mean()
plt.plot(gd.index,gd.values,marker='o',linestyle='-',color='red')
plt.title('Customer service calls vs Total day minutes',fontsize=13,style='italic')
plt.xlabel('Customer service calls',fontsize=10)
plt.ylabel('Total Day Minutes',fontsize=10)

plt.savefig('Line chart for Customer service calls vs Total day minutes .png',dpi=500,bbox_inches='tight')
plt.show()
```

#### INTERPRETATION:
- The line plot shows the average total day minutes vs Customer service calls .
- It is found that customers those who have made 0-4 service calls have medium day minutes.
- At the right side the graph increases suddenly which means that the customer have made more than
  7 customer service calls have the highest Total Day Minutes.
- The minimum number of Total Day Minutes has around 7 customer service calls.

#### c) SCATTER PLOT:

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=churn['Total intl minutes'],y=churn['Total intl charge'],hue=churn['Churn'])
plt.title('Scatter Plot for Total International Call Minutes vs Charges',color='Red')
plt.xlabel('Total International Minutes',fontsize=9)
plt.ylabel('Total International Charges',fontsize=9)
plt.legend(title='Churn Status')

plt.savefig('Scatter Plot for international calls minutes vs charges .png',dpi=500,bbox_inches='tight')
plt.show()
```

#### INTERPRETATION:
- The scatter plot is used to show the relationship between total international calls minutes vs charges.
- The plot has two colors: Orange-> Churn customers, Blue-> Unchurned customers.
- The plot shows a strong linear relationship means
            with increase in total international calls minutes, total international charges also increases.
- But these features couldn't clearly help in distinguishing churned and non churned customers.
