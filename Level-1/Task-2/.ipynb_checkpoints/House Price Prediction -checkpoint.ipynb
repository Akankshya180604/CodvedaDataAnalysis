## OBJECTIVE:
- Calculate summary statistics (mean, median, mode,standard deviation).
- Visualize data distributions using histograms, boxplots and scatter plots.
- Find correlations between numerical features.

### 1. IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. READING THE CSV FILE

```python
house=pd.read_csv(r'House Price Prediction DataSet.csv',header=None)
print(house)
```

### 3. RENAMING THE COLUMNS

```python
house.columns=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']
house.head()
```

### 4. CHECKING FOR NULL VALUES

```python
house.isnull().sum()
```

```python
house.describe()
```

```python
house['MEDV']=house['MEDV'].fillna(23.750442)
```

```python
house.isnull().sum()
```

### 5. VERIFYING DATA TYPES

```python
house.dtypes
```

### 6. SUMMARY STATISTICS

```python
numeric_col=house.select_dtypes(include=['number'])
```

#### a)MEAN: mean is the sum of all values divided by the number of values

```python
print('Mean:')
numeric_col.mean().round(2)
```

#### b)median: The median is the middle value when the values are sorted.

```python
numeric_col.median().round(2)
```

#### c)MODE: mode is the value that occurs most of the time in a sequence of numbers

```python
numeric_col.mode().iloc[0].round(2)
```

#### d. STANDARD DEVIATION: It measures the amount of variation from the mean.

```python
numeric_col.std().round(2)
```

### 7. VISUALIZING THE PREDICTION

#### a) HISTOGRAM

```python
plt.figure(figsize=(8,6))
sns.histplot(house['MEDV'],edgecolor='Black',kde=True)
plt.title('Histogram for house price prediction')
plt.xlabel('House price')
plt.ylabel('Frequency')
plt.show()
```

#### INTERPRETATION:
- This histogram represents the distribution of house prices.
- It reveals that the majority of house prices fall within the $18000 to $25000 range.
- The highest house price is approximately $50000 .
- Since most of the values are concentrated on the lower end of the scale, the distribution is right skewed.

#### b)BOXPLOT

```python
plt.figure(figsize=(8,6))
sns.boxplot(y=house['CRIM'])
plt.title('Box plot')
plt.ylabel('Crime rate')
plt.show()
```

#### INTERPRETATION:
- The box plot illustartes the distribution of crime rate across various towns.
- It clearly highlights the presense of numerous outliers in the dataset.
- Most of the outliers located on the higher end, indicating that certain towns have significantly higher crime rates compared to others.

#### c)SCATTER PLOT

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=house['MEDV'],y=house['CRIM'],hue=house['CHAS'])
plt.title('Scatter plot for House Price and crime rate')
plt.xlabel('House Price')
plt.ylabel('Crime rate')
plt.show()
```

#### INTERPRETATION:
- The scatter plot displays the relationship between 'MEDV'(House Price) and 'CRIM'(Crime Rate) with color variations indicating the proxy of Charles River.
- Lighter shades represent houses farther than the river while darker shades indicate those closer to it.
- The plot shows that areas with lower house price tend to have higher crime rates.
- Houses with higher prices are generally located in the area with lower crime rate.

### 8. CORRELATION BETWEEN NUMERICAL FEATURES

```python
print(house.corr(numeric_only=True))
```

```python
cm=house.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(cm,annot=True,cmap='Blues',fmt='.2f')
plt.title('Correlation for numerical features')
plt.show()
```

#### INTERPRETATION:
- The heatmap is used to identify the correlation between various features.
- It shows that 'RAD' and 'NOX' have a strong correlation of 0.99 indicating that areas with greater highway accessibility tend to experience higher levels of pollution.
