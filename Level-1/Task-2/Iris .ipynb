### 6. VISUALIZING DATA DISTRIBUTION

#### a)HISTOGRAM

```python
numeric_num=iris[['petal_length','petal_width','sepal_length','sepal_width']]
```

```python
plt.figure(figsize=(10,5))
iris['sepal_length'].hist(bins=20,grid=False,rwidth=0.9,edgecolor='black',color='blue')
plt.title('Histogram for sepal length')
plt.xlabel('Sepal length')
plt.ylabel('Frequency')
plt.show()
```

#### INTERPRETATION OF THE GRAPH:
- The histogram shows the distribution of sepal length of flowers.
- Here in the graph most of the data's are in between 5.0 c.m. to 6.5 c.m.
- The graph looks symmetrical so the data is evenly spread without skewness.
- The minimum sepal length is around 7.5 c.m and the maximum sepal length is around 5.0 c.m and 6.4 c.m also.

#### b) BOX PLOT

```python
sns.boxplot(x=iris['species'],y=iris['sepal_length'],data=iris)
plt.title('Box plot for Sepal length')
plt.ylabel('Sepal Length')
plt.show()
```

#### INTERPRETATION OF THE BOX PLOT:
- The box plot shows the distribution of sepal length across different species.
- As we can see setosa has shortest sepal length while Virginica has the longest sepal length.
- There are some outliers present in the virginica species.
- as we move from setosa-> versi color-> virginica , The sepal length incereases.

#### c) SCATTER PLOT

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=iris)
plt.title('Scatter plot for sepal length')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.grid(False)
plt.show()
```

##### 

#### INTERPRETATION FOR SCATTER PLOT:
- The scatter plot illustrates the relationship between sepal length and sepal width across various species.
- Each color corresponds a species
              Blue for Setosa, Orange for Versi color, Green for virginica
- As sepal length increases, the species tend to separate into distinct groups making identification easier.

#### EXTRA POINT:
- Versi color generally has longer sepal lengths but not wider sepal width.
- Setosa is easy to spot but versicolor and virginica look similar and overlap a little bit.

### 7. CORRELATION BETWEEN NUMERICAL FEATURES

```python
corr_matrix=iris.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix,annot=True,cmap='Blues',linewidth=0.5,fmt='.2f')
plt.title('Heatmap of numerical feature')
plt.show()
print('\n Correlation Matrix:')
print(corr_matrix)
```

#### INTERPRETATION OF HEATMAP:
- The heatmap displays how the numerical features are correlated with each other.
- Darker colors indicate a stronger positive correlation while the lighter shades shows a negative correlation.
- As petal width increase, petal length also increases.
- When petal length increases, sepal width tends to decreases.

#### EXTRA POINTS:
- There is very little to no relationship between sepal length and sepal width.
- Petal length and petal width show a strong correlation i.e. they increase together.
- Sepal length has a moderate correlation with the petal dimensions.
- Sepal width is weakly or even negatively related to the other features.
